# importing required packages

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import scipy.stats as stat
import numpy as np
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler
import pylab as pl
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,Dropout
from sklearn.metrics import *
```

```python
bank_node=pd.read_csv("BankNote_Authentication.csv")
```

```python
bank_node.columns
```

```python
bank_node["variance"].skew()
```

# Checking data balanced or not

```python
bank_node["class"].value_counts().plot.bar()
```

# distribution and skewness with Q-Q plot

```python
def plot_data(df,feature):
    print("Skewness of "+feature+" is",df[feature].skew())
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    pl.suptitle(feature)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pl)
    plt.show()
```

```python
for column in bank_node.iloc[:,:-1]:
    plot_data(bank_node,column)
```

# Spliting the data for train and test

```python
X=bank_node[['variance', 'skewness', 'curtosis', 'entropy']]
y=bank_node["class"]
```

```python
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.3,random_state=41)
```

# Standardization

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
```

```python
X_test=sc.transform(X_test)
```

# Model developing

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.0008,beta_1=0.95)
```

```python
model = Sequential()
model.add(Dense(100, input_dim=4, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,epochs=100,batch_size=32,validation_data=(X_test,y_test))
```

# ploting loss and accuracy for every epochs

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# Perfomence metrics

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred = (y_pred > 0.5)
```

### Test accuracy

```python
accuracy_score(y_test,y_pred)
```

### confusion_matrix for test data

```python
confusion_matrix(y_test,y_pred)
```

### f1_score for test data

```python
f1_score(y_test,y_pred)
```
